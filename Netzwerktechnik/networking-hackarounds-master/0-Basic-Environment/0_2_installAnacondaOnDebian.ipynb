# Anaconda Teaser
Anaconda is an Environment Management Center for Python Environments. Based on Anaconda you can create and manage environments

Environments have:
* an Python Version (e.g. 2.X or 3.X)
* indiviual packages installed
* indiviual Applications (Jupyter / I-Python etc) installed

Consequently if you write Python-Code in a certain Environment it will allways run in this Environment
