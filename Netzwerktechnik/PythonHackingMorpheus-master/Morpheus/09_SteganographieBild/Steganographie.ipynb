```python
from IPython.display import YouTubeVideo

def display_yotube_video(url, **kwargs):
    """
    Displays a Youtube video in a Jupyter notebook.
    
    Args:
        url (string): a link to a Youtube video.
        **kwargs: further arguments for IPython.display.YouTubeVideo
    
    Returns:
        YouTubeVideo: a video that is displayed in your notebook.
    """
    id_ = url.split("=")[-1]
    return YouTubeVideo(id_, **kwargs)

display_yotube_video("https://www.youtube.com/watch?v=PK3L62Vkux4", width=800, height=600)
```

```python
# Read
f = open("Snoopy.jpg", encoding = "ISO-8859-1", mode = "r") # mit dem write mode kann ich nicht lesen, daher hier neu einlesen

try:
    byte = f.read(1)
    str = "" # string den wir erstellen
    while byte != "": # in jpg ist leerer sting das ende des files
        str = str + byte
        byte = f.read(1)
    print(str) # Bild ausgeben
    
finally:
    f.close()
    
# Am ende des Outputs ist der input aus dem write
```
