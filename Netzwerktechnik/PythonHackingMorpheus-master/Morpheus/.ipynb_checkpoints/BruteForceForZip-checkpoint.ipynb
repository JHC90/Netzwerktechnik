```python
from IPython.display import YouTubeVideo

def display_yotube_video(url, **kwargs):
    """
    Displays a Youtube video in a Jupyter notebook.
    
    Args:
        url (string): a link to a Youtube video.
        **kwargs: further arguments for IPython.display.YouTubeVideo
    
    Returns:
        YouTubeVideo: a video that is displayed in your notebook.
    """
    id_ = url.split("=")[-1]
    return YouTubeVideo(id_, **kwargs)

display_yotube_video("https://www.youtube.com/watch?v=lyJfLUYQA5w", width=800, height=600)
```

```python
import zipfile
import itertools
import string
from threading import Thread
```

```python
# to create a zipfile just use 7-zip
zipFile = zipfile.ZipFile("./File2BsafedbyZip.zip")

# potential Sign space
myLetters = string.ascii_letters + string.digits + string.punctuation
```

```python
# Simple Brute-Force FCT
def crack(zip, pwd):
    try:
        zip.extractall(pwd=str.encode(pwd))
        print("sucess_ password is: " + pwd)
    except Exception:
        pass
```

```python
for i in range(3,10):
    # hier muss ein generator her, da wenn die liste erstellt wird, sprengts alles
    for j in map(''.join, itertools.product(myLetters, repeat=3)):
        t=Thread(target=crack, args=(zipFile, j))
        t.start()  
```
